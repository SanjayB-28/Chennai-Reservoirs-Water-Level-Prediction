# STEP 2: Build the `flask` module
The `flask` module gets date from website, and generates optimal outflow for the given date using the **ANN model**.

```python
#MIN-MAX NORMALISATION
def normaliser(data,sdate):
    x=datetime.strptime(sdate,'%d/%m/%Y')
    y=x-timedelta(days=5)
    bdate=y.strftime("%d")+'/'+y.strftime("%m")+'/'+y.strftime("%Y")
    y=x+timedelta(days=35)
    edate=y.strftime("%d")+'/'+y.strftime("%m")+'/'+y.strftime("%Y")
    
    siro=data.loc[bdate:edate]
    
    siro=siro.reset_index(drop=True)
    
    siro=siro.drop(columns=['nth day in year'])
    return siro

def sequence_prepare(raw,lookback=5):
    sequence=np.zeros((30,5,raw.shape[-1]))
    for i in range(lookback,lookback+30):
        for j in range(0,lookback):
            sequence[i-lookback][j]=raw.loc[i+j-lookback,:]
        
    return sequence



def inputmaker(sdate):
    siro=[]
    sequence=[]
    for i in range(0,4):
        siro.append(normaliser(dianna[i],sdate))
        #print(siro[i].describe())
        sequ=sequence_prepare(siro[i])
        sequence.append(sequ)
    return sequence

def validator(sdate):
    ssdate=sdate;
    x=datetime.strptime(ssdate,'%d/%m/%Y')
    y=x-timedelta(days=365*3)
    sdate=y.strftime("%d")+'/'+y.strftime("%m")+'/'+y.strftime("%Y")
    lisres=list()
    signalflag=0
    dateres=list()

 

    for i in range(30):
        sequence=inputmaker(sdate)
        x=datetime.strptime(ssdate,'%d/%m/%Y')
        y=x+timedelta(days=1)
        ssdate=y.strftime("%d")+'/'+y.strftime("%m")+'/'+y.strftime("%Y")
        dateres.append(ssdate)
        
        x=datetime.strptime(sdate,'%d/%m/%Y')
        y=x+timedelta(days=1)
        sdate=y.strftime("%d")+'/'+y.strftime("%m")+'/'+y.strftime("%Y")
        starter=0
        i=0
        rans=modelv2.predict([sequence[0][i:i+1],sequence[1][i:i+1],sequence[2][i:i+1],sequence[3][i:i+1]])
        r=[rans[0][0][0],rans[1][0][0],rans[2][0][0],rans[3][0][0]]
        print(r)
        #Ceil function: to round outflow values   
        def ceil(x):
            if(x-x//1>0):
                return x//1+1
            else:
                return x

 


        #to prepare a return sequence : from [a,b,c,d] model output to "a,b,c,d" string for the website
        
        for i in range(4):
            lisres.append(ceil(r[i]))
            
    for dater in dateres:
        lisres.append(dater)
    lisres.append(str(signalflag))
    return lisres


storage_cap={
    0:3300,
    1:3231,
    2:881,
    3:3645}

def isflood(i,val):
    if(val>=0.6*storage_cap[i]):
        return True;
    else:
        return False;

def isdrought(i,val):
    if(val<=0.1*storage_cap[i]):
        return True;
    else:
        return False;
```

```python
app = Flask(__name__)
content=''


@app.route('/')
def index():
    return docs



@app.route('/_cal')
def login():
    resultino="";
    sdate = request.args.get('nm')
    lisres=list()
    signalflag=0
    dateres=list()
    if(sdate[len(sdate)-1]=='1' and sdate[len(sdate)-2]=='2'):
        lisres=validator(sdate)
        for i in lisres:
            resultino+=(","+str(i))
        print(resultino)
        return resultino
    
        
    for i in range(30):
        sequence=inputmaker(sdate)
        x=datetime.strptime(sdate,'%d/%m/%Y')
        y=x+timedelta(days=1)
        sdate=y.strftime("%d")+'/'+y.strftime("%m")+'/'+y.strftime("%Y")
        dateres.append(sdate)
        starter=0
        i=0
        rans=modelv2.predict([sequence[0][i:i+1],sequence[1][i:i+1],sequence[2][i:i+1],sequence[3][i:i+1]])
        r=[rans[0][0][0],rans[1][0][0],rans[2][0][0],rans[3][0][0]]
        print(r)
        #Ceil function: to round outflow values   
        def ceil(x):
            if(x-x//1>0):
                return x//1+1
            else:
                return x


        #to prepare a return sequence : from [a,b,c,d] model output to "a,b,c,d" string for the website
        
        for i in range(4):
            lisres.append(ceil(r[i]))
            if(isflood(i,r[i])):
                signalflag=1
            if(isdrought(i,r[i])):
                signalflag=2
    for dater in dateres:
        lisres.append(dater)
    lisres.append(str(signalflag))
    print('Output passed: ',lisres)
    resultino="";
    for i in lisres:
        resultino+=(","+str(i))
    print(resultino)
    return resultino



if __name__ == '__main__':
       app.run()
```
